## 1. What is Git? What is GitHub?

| | **Git** | **GitHub** |
|---|---|---|
| What it is | A **version control system** (software) | A **website/service** that hosts Git repositories |
| Runs where | Locally on your machine | In the cloud |
| Purpose | Track changes to files over time | Share, back up, and collaborate on Git repos |
| Analogy | The engine | The garage where cars (repos) are parked and shared |

**Key vocabulary:**
- **Repository (repo):** A folder tracked by Git, containing your project + its full history.
- **Commit:** A saved snapshot of your project at a point in time.
- **Branch:** An independent line of development (e.g. `main`, `feature/login`).
- **Remote:** A version of your repo hosted elsewhere (e.g. on GitHub), usually named `origin`.
- **Clone:** Downloading a copy of a remote repo to your machine.
- **Push / Pull:** Sending your commits to a remote / fetching commits from a remote.
- **Merge / Pull Request (PR):** Combining changes from one branch (or fork) into another.

## 4. The Core Workflow: Add → Commit

Git has three main areas:

```
Working Directory  --git add-->  Staging Area  --git commit-->  Repository (History)
   (your files)                   (what will be                 (permanent snapshots)
                                    in next commit)
```

Let's create a file, stage it, and commit it.


````python
# Create a simple file using Python's %%writefile magic
%%writefile README.md
# My First Repo

This is a test project to learn Git and GitHub.
````

````python
# See that Git notices the new file as 'untracked'
!git status
````

````python
# Stage the file (add it to the staging area)
!git add README.md

# Alternative: stage EVERYTHING that's changed
# !git add .

!git status
````

````python
# Commit the staged changes with a descriptive message
!git commit -m "Initial commit: add README"

# View the commit history
!git log
````

````python
# A more compact, readable log view
!git log --oneline --graph --all --decorate
````

**Best practice for commit messages:**
- Keep the first line under ~50 characters, written in the imperative mood: `"Add login form"` not `"Added login form"` or `"Adds login form"`.
- Add more detail in the body if needed (a blank line after the summary, then details).

## 6. Branching & Merging

Branches let you work on new features without touching your stable `main` branch.

```
main:      A---B---C---------F
                    \        /
feature:             D------E
```


````python
# List existing branches (the * marks the current branch)
!git branch

# Create AND switch to a new branch in one step
!git checkout -b feature/add-license
````

````python
%%writefile LICENSE.txt
MIT License

Copyright (c) 2026 Your Name

Permission is hereby granted, free of charge, to any person obtaining a copy...
````

````python
!git add LICENSE.txt
!git commit -m "Add MIT license"
!git log --oneline --graph --all
````

````python
# Switch back to main
!git checkout main

# Merge the feature branch into main
!git merge feature/add-license -m "Merge feature/add-license into main"

!git log --oneline --graph --all
````

````python
# Once merged, you can safely delete the feature branch
!git branch -d feature/add-license
!git branch
````

**Common branch commands:**
```bash
git branch                 # list local branches
git branch -a               # list all branches (local + remote)
git checkout -b <name>       # create + switch to new branch
git checkout <name>          # switch to existing branch
git switch <name>            # modern alternative to checkout
git branch -d <name>         # delete a merged branch
git branch -D <name>         # force-delete an unmerged branch
```

## 8. Undoing Things

Mistakes happen — Git gives you several tools to fix them, depending on the situation.

| Situation | Command |
|---|---|
| Unstage a file (keep changes) | `git restore --staged <file>` |
| Discard changes in a file (dangerous!) | `git restore <file>` |
| Change the last commit's message/content | `git commit --amend` |
| Move branch pointer, keep history as new commit (safe, for shared branches) | `git revert <commit>` |
| Move branch pointer back, rewriting history (only for local/unshared work) | `git reset --soft/--mixed/--hard <commit>` |


````python
# Example: stage a file, then unstage it
%%writefile scratch.txt
temporary notes
````

````python
!git add scratch.txt
!git status
````

````python
# Unstage it (file stays on disk, just not staged)
!git restore --staged scratch.txt
!git status
````

````python
# Clean up the scratch file entirely
!rm scratch.txt
````

````python
# Example of git revert: safely undoes a commit by creating a NEW commit that
# reverses it (does not rewrite history) - safe for branches others are using.
# git revert <commit-hash>

# Example of git reset (rewrites history - use ONLY on local, unpushed commits)
# git reset --soft HEAD~1   # undo last commit, keep changes staged
# git reset --mixed HEAD~1  # undo last commit, keep changes unstaged (default)
# git reset --hard HEAD~1   # undo last commit AND discard changes (DESTRUCTIVE)

print("These are shown as comments above — read carefully before ever running --hard resets!")
````

## 10. Connecting to GitHub

To push code to GitHub from Colab, you need to authenticate. GitHub no longer accepts your account password for Git operations — you need a **Personal Access Token (PAT)** (or SSH keys).

### Step A — Create a Personal Access Token (do this on github.com, not in Colab)
1. Go to **github.com → Settings → Developer settings → Personal access tokens → Tokens (classic)** (or "Fine-grained tokens").
2. Click **Generate new token**.
3. Give it the `repo` scope (full control of private repositories) at minimum.
4. Copy the generated token — **you will only see it once**.

### Step B — Store it safely in Colab
Never hardcode your token in a notebook cell you plan to share! Use Colab's **Secrets** manager (the 🔑 key icon in the left sidebar) or the input prompt below, which hides your typing.


````python
import getpass

# This creates a hidden input box in the cell output — your token won't be shown or saved in the notebook
github_username = input("GitHub username: ")
github_token = getpass.getpass("GitHub Personal Access Token: ")

print("Credentials captured for this session (not saved to disk).")
````

### Step C — Create a repository on GitHub
Go to **github.com → New repository**, give it a name (e.g. `my-first-repo`), and **do not** initialize it with a README (since we already have local commits) — or do, and we'll handle the merge.


````python
# Set your repo URL (replace 'my-first-repo' with the actual repo name you created)
repo_name = "my-first-repo"
remote_url = f"https://{github_username}:{github_token}@github.com/{github_username}/{repo_name}.git"

# Add GitHub as a remote named 'origin'
!git remote add origin {remote_url}

# Confirm the remote was added (this will show the URL WITHOUT the token for safety in modern git)
!git remote -v
````

⚠️ **Security warning:** The command above embeds your token directly in the remote URL for simplicity in this demo. For real projects, prefer:
- **SSH keys** (see Section 10b below), or
- The [GitHub CLI](https://cli.github.com/) (`gh auth login`), or
- Git's [credential helper](https://git-scm.com/docs/gitcredentials) so tokens aren't stored in plain text in `.git/config`.

Also: **revoke any token** you typed into a shared or public notebook once you're done testing.


### 10b. Alternative: Authenticating with SSH keys

## 13. Fetch vs. Pull vs. Push — What's the Difference?

| Command | What it does |
|---|---|
| `git fetch` | Downloads new commits from the remote but **doesn't** merge them into your current branch. Safe to run anytime. |
| `git pull` | Runs `git fetch` **then** merges (or rebases) those changes into your current branch. |
| `git push` | Uploads your local commits to the remote. |

**Recommended habit:** `git fetch` + `git log origin/main --oneline` to *see* what's new before blindly pulling, especially on shared/important branches.


````python
%cd /content/my_first_repo

# Fetch remote changes without merging
!git fetch origin

# See what's different between your local main and origin's main
!git log main..origin/main --oneline   # commits on origin not yet in your local main
!git log origin/main..main --oneline   # commits local that origin doesn't have yet
````

````python
# Pull = fetch + merge in one step
!git pull origin main
````

## 14. Collaboration: Forks & Pull Requests

This is the standard open-source contribution workflow on GitHub:

1. **Fork** the repository (creates your own copy under your GitHub account, via the website's "Fork" button).
2. **Clone** your fork locally: `git clone git@github.com:<you>/<repo>.git`
3. Add the original repo as a second remote, conventionally called `upstream`:
   ```bash
   git remote add upstream https://github.com/<original-owner>/<repo>.git
   ```
4. Create a feature branch: `git checkout -b fix/typo-in-docs`
5. Make changes, commit, and push to **your fork**: `git push origin fix/typo-in-docs`
6. On GitHub, click **"Compare & pull request"** to open a **Pull Request (PR)** from your branch into the original repo's `main`.
7. Respond to review comments by committing more changes and pushing again — the PR updates automatically.
8. Once approved, a maintainer **merges** your PR. 🎉
9. Keep your fork updated over time:
   ```bash
   git fetch upstream
   git checkout main
   git merge upstream/main
   git push origin main
   ```

## 15. Resolving Merge Conflicts

A conflict happens when Git can't automatically combine changes — usually because two branches edited the **same lines** of the **same file** differently. Let's simulate one.


````python
%cd /content/my_first_repo

# Make sure we're on main and up to date
!git checkout main

# Branch A: edits the title
!git checkout -b branch-a
!sed -i '1s/.*/# My First Repo (Branch A Title)/' README.md
!git commit -am "Branch A: change title"

# Back to main, Branch B: edits the SAME line differently
!git checkout main
!git checkout -b branch-b
!sed -i '1s/.*/# My First Repo (Branch B Title)/' README.md
!git commit -am "Branch B: change title differently"

!git log --oneline --graph --all
````

````python
# Merge branch-a into main first — this works cleanly
!git checkout main
!git merge branch-a -m "Merge branch-a"
!head -n 1 README.md
````

````python
# Now merge branch-b into main — CONFLICT! Both changed line 1 differently.
!git merge branch-b -m "Merge branch-b" || echo "--- CONFLICT DETECTED (expected) ---"
!git status
````

````python
# Look at the conflict markers Git inserted into the file
!cat README.md
````

You'll see something like:
```
<<<<<<< HEAD
# My First Repo (Branch A Title)
=======
# My First Repo (Branch B Title)
>>>>>>> branch-b
```
- Everything between `<<<<<<< HEAD` and `=======` is **your current branch's** version.
- Everything between `=======` and `>>>>>>> branch-b` is the **incoming branch's** version.

**To resolve:** manually edit the file to keep what you want and delete the `<<<<<<<`, `=======`, `>>>>>>>` marker lines, then stage and commit.


````python
%%writefile README.md
# My First Repo (Merged Title — combined decision)

This is a test project to learn Git and GitHub.

## Goals
- Learn git add, commit, status
- Learn branching
- Push to GitHub
````

````python
# Mark the conflict as resolved by staging the file, then complete the merge commit
!git add README.md
!git commit -m "Resolve merge conflict between branch-a and branch-b"

!git log --oneline --graph --all
````

````python
# Clean up demo branches
!git branch -d branch-a
!git branch -d branch-b
!git branch
````

## 18. 📋 Git & GitHub Cheat Sheet

### Setup
```bash
git config --global user.name "Name"
git config --global user.email "email@example.com"
```

### Starting a repo
```bash
git init                 # start a new repo here
git clone <url>            # copy an existing remote repo
```

### Everyday workflow
```bash
git status                 # what's changed?
git add <file>               # stage a file
git add .                     # stage everything
git commit -m "message"        # save a snapshot
git log --oneline --graph        # view history
git diff                           # see unstaged changes
```

### Branching
```bash
git branch                     # list branches
git checkout -b <name>           # create + switch
git switch <name>                  # switch branches
git merge <name>                     # merge into current branch
git branch -d <name>                   # delete merged branch
```

### Remote / GitHub
```bash
git remote add origin <url>       # link to GitHub repo
git push -u origin main             # first push, sets upstream
git push                              # push after that
git pull                                # fetch + merge
git fetch                                 # download without merging
```

### Undo
```bash
git restore --staged <file>     # unstage
git restore <file>                # discard local changes (careful!)
git commit --amend                  # edit last commit
git revert <commit>                   # safely undo a commit (new commit)
git reset --hard <commit>               # rewrite history (local only, careful!)
```

### Stash
```bash
git stash          # save uncommitted work
git stash pop        # bring it back
git stash list         # see all stashes
```

### Tags
```bash
git tag -a v1.0 -m "msg"    # annotated tag
git push origin --tags        # push all tags
```
