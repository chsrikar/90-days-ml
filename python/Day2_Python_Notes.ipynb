## 1. Conditional Statements — `if / elif / else`

Control flow means **making decisions** in code.  
Python uses indentation (4 spaces) to define blocks — there are no `{}` like in C/Java.

```python
if condition:
    # runs if condition is True
elif another_condition:
    # runs if first was False, this is True
else:
    # runs if ALL above were False
```

> ⚠️ **Indentation is NOT optional in Python.** A wrong indent = an error or silent wrong behaviour.

````python
# ── 1.1 Basic if / elif / else ──────────────────────────────
marks = 82

if marks >= 90:
    grade = 'A'
elif marks >= 75:
    grade = 'B'
elif marks >= 60:
    grade = 'C'
elif marks >= 40:
    grade = 'D'
else:
    grade = 'F'

print(f'Marks: {marks} → Grade: {grade}')

# Try changing marks to 45, 91, 30 and re-run
````

````python
# ── 1.2 Nested if — if inside another if ────────────────────
age       = 20
has_id    = True

if age >= 18:
    if has_id:
        print('✅ Entry allowed — age valid + ID present')
    else:
        print('❌ Entry denied — no ID')
else:
    print('❌ Entry denied — underage')

# Same logic using and (cleaner for simple cases)
if age >= 18 and has_id:
    print('✅ (using and) Entry allowed')
````

````python
# ── 1.3 Ternary (one-line if) ───────────────────────────────
# Syntax: value_if_true  if  condition  else  value_if_false
# You've seen this in f-strings already!

marks = 73
result = 'Pass' if marks >= 40 else 'Fail'
print(result)

# Chained ternary (use carefully — readability matters)
grade = 'A' if marks >= 90 else 'B' if marks >= 75 else 'C' if marks >= 60 else 'F'
print(grade)

# Bigger number
a, b = 15, 27
bigger = a if a > b else b
print(f'Bigger of {a} and {b} is {bigger}')
````

````python
# ── 1.4 if with logical operators ───────────────────────────
cgpa      = 8.5
backlogs  = 0
attendance= 80

# Placement eligibility check
if cgpa >= 7.0 and backlogs == 0 and attendance >= 75:
    print('✅ Eligible for campus placements')
else:
    if cgpa < 7.0:
        print('❌ CGPA below cutoff')
    if backlogs > 0:
        print('❌ Active backlogs')
    if attendance < 75:
        print('❌ Low attendance')
````

````python
# ── 1.5 Truthy and Falsy values ─────────────────────────────
# In Python, these values are FALSY (behave like False in if):
# 0, 0.0, '', [], {}, None, False
# Everything else is TRUTHY

values = [0, 1, '', 'hello', None, [], [1,2], False, True, -1]

for v in values:                   # we'll learn for loops next!
    print(f'{str(v):10} → {"Truthy" if v else "Falsy"}')
````

## 2. `while` Loop — Repeat Until Condition is False

Use `while` when you **don't know in advance** how many times to repeat.

```python
while condition:
    # body — runs as long as condition is True
    # MUST update something so condition eventually becomes False!
```

> ⚠️ If the condition never becomes `False` → **infinite loop**. Use `Ctrl+C` to stop or Runtime → Interrupt in Colab.

````python
# ── 2.1 Basic while loop ─────────────────────────────────────
count = 1

while count <= 5:
    print(f'Count: {count}')
    count += 1          # MUST increment — otherwise infinite loop!

print('Loop finished. count =', count)
````

````python
# ── 2.2 while with user input — repeat until correct ────────
secret = '1234'
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    guess = input(f'Enter PIN (attempt {attempts+1}/{max_attempts}): ')
    attempts += 1
    if guess == secret:
        print('✅ Access granted!')
        break           # exit loop immediately
else:
    # the while-else runs only if loop completed WITHOUT break
    print('❌ Too many wrong attempts. Account locked.')
````

````python
# ── 2.3 while with accumulator pattern ──────────────────────
# Keep taking numbers from user until 0 is entered
# Then print the sum and count

total   = 0
count   = 0

print('Enter numbers (0 to stop):')
while True:                            # intentionally infinite
    num = int(input('Number: '))
    if num == 0:
        break                          # THIS is the exit condition
    total += num
    count += 1

if count > 0:
    print(f'\nCount   : {count}')
    print(f'Total   : {total}')
    print(f'Average : {total/count:.2f}')
else:
    print('No numbers entered.')
````

````python
# ── 2.4 Countdown timer (classic while pattern) ──────────────
n = 10

print('Countdown:')
while n > 0:
    print(n, end=' ')
    n -= 1

print('\n🚀 Liftoff!')
````

## 3. `for` Loop — Iterate Over a Sequence

Use `for` when you **know exactly what to iterate over** — a range, list, string, etc.

```python
for variable in sequence:
    # body
```

> 💡 `for` loops in Python iterate over **items**, not indices (unlike C's `for(i=0; i<n; i++)`)  
> This makes them cleaner and less error-prone.

````python
# ── 3.1 for over a range ─────────────────────────────────────
# range(stop)          → 0, 1, 2 ... stop-1
# range(start, stop)   → start, start+1 ... stop-1
# range(start,stop,step) → start, start+step ... stop-1

print('range(5)      →', end=' ')
for i in range(5):
    print(i, end=' ')

print('\nrange(1,6)    →', end=' ')
for i in range(1, 6):
    print(i, end=' ')

print('\nrange(0,10,2) →', end=' ')
for i in range(0, 10, 2):
    print(i, end=' ')

print('\nrange(5,0,-1) →', end=' ')
for i in range(5, 0, -1):           # countdown!
    print(i, end=' ')
print()
````

````python
# ── 3.2 for over a string ────────────────────────────────────
# A string is a sequence of characters
name = 'Srikar'

print('Characters in name:')
for char in name:
    print(char, end=' → ')

print('\n\nVowel count:')
vowels = 'aeiouAEIOU'
count = 0
for char in name:
    if char in vowels:
        count += 1
print(f'  "{name}" has {count} vowel(s)')
````

````python
# ── 3.3 for over a list ──────────────────────────────────────
subjects = ['Python', 'Math', 'Physics', 'DSA', 'ML']

print('My subjects:')
for subject in subjects:
    print(f'  • {subject}')

# With enumerate — get index AND value at the same time
print('\nWith index:')
for i, subject in enumerate(subjects, start=1):
    print(f'  {i}. {subject}')
````

````python
# ── 3.4 Nested for loops ─────────────────────────────────────
# A loop inside another loop
# Outer runs n times, inner runs m times → total n×m iterations

# Multiplication table
print('Multiplication table (1–5):')
print('-' * 30)
for i in range(1, 6):
    for j in range(1, 6):
        print(f'{i*j:3}', end=' ')
    print()              # newline after each row
````

````python
# ── 3.5 Pattern printing with nested loops ───────────────────
# Classic beginner exercise — great for loop intuition

n = 5

print('Right-angle triangle:')
for i in range(1, n+1):
    print('* ' * i)

print('\nInverted triangle:')
for i in range(n, 0, -1):
    print('* ' * i)

print('\nNumber pyramid:')
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=' ')
    print()
````

````python
# ── 3.6 for with else ────────────────────────────────────────
# The else block runs ONLY if the loop completed WITHOUT break
# Very useful for search scenarios

nums = [3, 7, 12, 5, 9, 21, 4]
target = 10

for num in nums:
    if num == target:
        print(f'✅ Found {target} in the list!')
        break
else:
    print(f'❌ {target} not found in the list')

# Change target to 12 to see it found
````

## 4. `range()` — Deep Dive

`range()` generates a sequence of integers. It's **lazy** — doesn't create a list in memory, generates one value at a time.

```python
range(stop)              # 0 to stop-1
range(start, stop)       # start to stop-1
range(start, stop, step) # start to stop-1, jumping by step
```

````python
# ── 4.1 range() variations ───────────────────────────────────

# Convert to list to see all values at once
print(list(range(10)))              # [0,1,2,...,9]
print(list(range(1, 11)))           # [1,2,3,...,10]
print(list(range(0, 20, 3)))        # [0,3,6,9,12,15,18]
print(list(range(10, 0, -1)))       # [10,9,8,...,1]  countdown
print(list(range(0, -10, -2)))      # [0,-2,-4,-6,-8]

# Length of a range
print(len(range(5, 50, 5)))         # how many steps? 9
````

````python
# ── 4.2 Common range patterns in ML/DSA ─────────────────────

# Sum of first n numbers (Gauss formula check)
n = 100
total = sum(range(1, n+1))
gauss = n * (n + 1) // 2
print(f'Sum 1 to {n} = {total} | Gauss formula = {gauss} | Match: {total==gauss}')

# Sum of squares
sum_sq = sum(i**2 for i in range(1, 6))  # generator expression preview
print(f'1² + 2² + 3² + 4² + 5² = {sum_sq}')

# Generate even numbers up to 20
evens = list(range(0, 21, 2))
print('Evens up to 20:', evens)
````

## 5. `break`, `continue`, `pass` — Loop Control

| Keyword | What it does |
|---|---|
| `break` | **Exit the loop immediately** — skip remaining iterations |
| `continue` | **Skip current iteration** — jump to next iteration |
| `pass` | **Do nothing** — placeholder so Python doesn't complain about empty blocks |

````python
# ── 5.1 break — exit when condition is met ───────────────────
print('Finding first number divisible by 7 above 20:')

for n in range(21, 100):
    if n % 7 == 0:
        print(f'  Found: {n}')
        break              # stop as soon as found

# break in while
print('\nSecurity check:')
password = 'iedc2025'
attempts = 0
while True:
    inp = input('Password: ')
    attempts += 1
    if inp == password:
        print(f'✅ Correct! ({attempts} attempt(s))')
        break
    elif attempts >= 3:
        print('❌ Locked out after 3 attempts')
        break
    else:
        print(f'Wrong. {3 - attempts} attempt(s) left.')
````

````python
# ── 5.2 continue — skip current iteration ────────────────────

# Print only odd numbers from 1–10
print('Odd numbers:')
for i in range(1, 11):
    if i % 2 == 0:
        continue           # skip evens
    print(i, end=' ')

# Skip negative numbers in a list
print('\n\nPositive numbers only:')
data = [3, -1, 7, -4, 0, 9, -2, 5]
for num in data:
    if num <= 0:
        continue
    print(num, end=' ')
````

````python
# ── 5.3 pass — placeholder (used in incomplete code) ─────────

# Used when Python syntactically requires a block,
# but you haven't written the logic yet

for i in range(5):
    if i == 3:
        pass               # TODO: handle this case later
    else:
        print(i, end=' ')

# Also used in empty functions / classes (you'll use this in OOPs week)
def placeholder_function():
    pass                   # function exists but does nothing yet
````

````python
# ── 5.4 break vs continue — side by side comparison ──────────
print('=== break stops the loop ===')
for i in range(10):
    if i == 5:
        break
    print(i, end=' ')     # prints 0 1 2 3 4

print('\n\n=== continue skips just that iteration ===')
for i in range(10):
    if i == 5:
        continue
    print(i, end=' ')     # prints 0 1 2 3 4 6 7 8 9 (5 is skipped)
````

## 6. Putting It All Together — Real Examples

````python
# ── 6.1 FizzBuzz (Classic interview question) ─────────────────
# Rules: 1 to 30
# Divisible by 3 → print Fizz
# Divisible by 5 → print Buzz
# Divisible by both → print FizzBuzz
# Otherwise → print the number

for i in range(1, 31):
    if i % 15 == 0:        # check 15 FIRST (both 3 and 5)
        print('FizzBuzz', end=' ')
    elif i % 3 == 0:
        print('Fizz', end=' ')
    elif i % 5 == 0:
        print('Buzz', end=' ')
    else:
        print(i, end=' ')
````

````python
# ── 6.2 Prime number check ────────────────────────────────────
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):   # only check up to √n
        if n % i == 0:
            return False
    return True

# Print all primes up to 50
primes = [n for n in range(2, 51) if is_prime(n)]
print('Primes up to 50:', primes)
print(f'Count: {len(primes)}')
````

````python
# ── 6.3 Number guessing game ──────────────────────────────────
import random

secret   = random.randint(1, 20)
attempts = 0
max_att  = 5

print('🎮 Guess the number between 1 and 20')
print(f'You have {max_att} attempts\n')

while attempts < max_att:
    guess    = int(input(f'Attempt {attempts+1}: Enter your guess: '))
    attempts += 1

    if guess == secret:
        print(f'\n🎉 Correct! You got it in {attempts} attempt(s)!')
        break
    elif guess < secret:
        remaining = max_att - attempts
        hint = '🔼 Too low'
    else:
        remaining = max_att - attempts
        hint = '🔽 Too high'

    if attempts < max_att:
        print(f'{hint}. {remaining} attempt(s) left.')
else:
    print(f'\n💀 Out of attempts! The number was {secret}')
````

````python
# ── 6.4 Marks analyser — realistic ML-adjacent example ────────
print('📊 Marks Analyser — enter marks for 5 subjects')
print('(Enter -1 to stop early)\n')

subjects = ['Python', 'Math', 'Physics', 'DSA', 'ML']
marks    = []

for subject in subjects:
    while True:                            # keep asking until valid input
        try:
            m = int(input(f'  {subject}: '))
            if m == -1:
                break
            if 0 <= m <= 100:
                marks.append(m)
                break
            else:
                print('  ⚠️ Enter a value between 0 and 100')
        except ValueError:
            print('  ⚠️ Enter a valid integer')
    if m == -1:
        break

if marks:
    total   = sum(marks)
    average = total / len(marks)
    highest = max(marks)
    lowest  = min(marks)
    grade   = 'A' if average >= 90 else 'B' if average >= 75 else 'C' if average >= 60 else 'F'

    print('\n' + '='*35)
    print('  📋 RESULT CARD')
    print('='*35)
    for i, (s, m) in enumerate(zip(subjects[:len(marks)], marks)):
        bar = '█' * (m // 10)
        print(f'  {s:8}: {m:3}/100  {bar}')
    print('='*35)
    print(f'  Total   : {total}')
    print(f'  Average : {average:.2f}')
    print(f'  Highest : {highest}')
    print(f'  Lowest  : {lowest}')
    print(f'  Grade   : {grade}')
    print('='*35)
````
