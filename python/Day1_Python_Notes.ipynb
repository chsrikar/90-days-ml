## 2. Variables

A **variable** is a named container that stores a value in memory.  
In Python — no type declaration needed. Python figures it out automatically.

```
variable_name = value
```

````python
# ── 2.1 Creating Variables ──────────────────────────────────
name    = 'Srikar'       # string
age     = 21             # integer
cgpa    = 8.7            # float
is_ceo  = True           # boolean

print(name)
print(age)
print(cgpa)
print(is_ceo)
````

### 2.2 Variable Naming Rules

- ✅ Must **start with a letter or underscore** `_` → `_name`, `name1` are valid
- ✅ Can contain letters, digits, underscores
- ✅ **Case-sensitive** → `name ≠ Name ≠ NAME`
- ❌ Cannot be a Python keyword → `if`, `for`, `while`, `class`, `def`…

> ⚠️ **Never name a variable** `list`, `str`, `type`, or `input` — these are Python built-in functions. You'll silently break them.

````python
# ── 2.3 Multiple Assignment (Pythonic tricks) ───────────────

# Unpack multiple values at once
x, y, z = 10, 20, 30
print(x, y, z)

# Assign same value to multiple variables
a = b = c = 0
print(a, b, c)

# Swap WITHOUT a temp variable (very Pythonic)
p, q = 5, 99
print('Before swap:', p, q)
p, q = q, p
print('After swap: ', p, q)
````

## 3. Data Types

Every value in Python has a **type**. The 5 core types for Day 1:

| Type | Keyword | Example | When you use it |
|---|---|---|---|
| Integer | `int` | `42, -7, 0` | Counts, indices, whole numbers |
| Float | `float` | `3.14, -0.5, 1e-3` | Decimal values, ML weights |
| String | `str` | `'hello'`, `"world"` | Text, names, labels |
| Boolean | `bool` | `True, False` | Flags, conditions, comparisons |
| NoneType | `None` | `None` | Absence of value, uninitialised vars |

````python
# ── 3.1 Checking the type of any variable ──────────────────
print(type(42))          # <class 'int'>
print(type(3.14))        # <class 'float'>
print(type('Srikar'))    # <class 'str'>
print(type(True))        # <class 'bool'>
print(type(None))        # <class 'NoneType'>
````

````python
# ── 3.2 Type Conversion (Casting) ──────────────────────────
# You can convert between types explicitly

print(int('42'))         # string → int  : 42
print(float(7))          # int → float   : 7.0
print(str(100))          # int → string  : '100'
print(bool(0))           # 0 is falsy    : False
print(bool(1))           # non-zero truthy: True
print(bool(''))          # empty string  : False
print(bool('hello'))     # non-empty     : True

# In ML you'll cast dtypes constantly!
# e.g. converting a string column → float before feeding to a model
````

````python
# ── 3.3 Strings — Key Operations ────────────────────────────
name = 'srikar chinthapenta'

print(name.upper())              # ALL CAPS
print(name.title())              # Title Case
print(name.replace('a', '@'))    # replace chars
print(len(name))                 # length of string
print(name[0])                   # first character  (indexing)
print(name[-1])                  # last character   (negative index)
print(name[0:6])                 # 'srikar'          (slicing)
print(name.split(' '))           # ['srikar', 'chinthapenta']
print('srikar' in name)          # True  (membership check)
````

````python
# ── 3.4 f-strings — Modern Python (MUST KNOW) ───────────────
# f-strings are the cleanest way to embed variables in text

name = 'Srikar'
role = 'CEO, IEDC VISAT'
cgpa = 8.7321

# Basic f-string
print(f'My name is {name} and I am {role}')

# Format float to 2 decimal places
print(f'CGPA: {cgpa:.2f}')

# Expression inside f-string
marks = 87
print(f'Marks: {marks}/100 → Percentage: {marks}%')
print(f'Pass or Fail: {"Pass" if marks >= 40 else "Fail"}')
````

## 4. Input & Output (I/O)

````python
# ── 4.1 print() — All the ways to print ────────────────────
print('Hello, World!')
print('Name:', 'Srikar', '| Age:', 21)    # multiple args (space-separated)
print('A', 'B', 'C', sep='-')             # custom separator → A-B-C
print('Line 1\nLine 2')                   # \n = newline
print('Loading...', end=' ')              # no newline at end
print('Done!')                             # continues on same line
````

````python
# ── 4.2 input() — Taking User Input ────────────────────────
# IMPORTANT: input() ALWAYS returns a STRING
# You must cast it if you need a number!

name  = input('Enter your name: ')         # string — fine as is
age   = int(input('Enter your age: '))     # cast to int
cgpa  = float(input('Enter your CGPA: ')) # cast to float

print(f'\nHello {name}!')
print(f'Age: {age} | CGPA: {cgpa:.2f}')
````

````python
# ── 4.3 Mini I/O Program ────────────────────────────────────
name  = input('Your name: ')
marks = int(input('Your marks out of 100: '))

grade = 'A' if marks >= 90 else 'B' if marks >= 75 else 'C' if marks >= 60 else 'D'

print(f'\nHello {name}!')
print(f'Marks  : {marks}/100')
print(f'Grade  : {grade}')
print(f'Result : {"Pass ✅" if marks >= 40 else "Fail ❌"}')
````

## 5. Operators

### 5.1 Arithmetic Operators

| Operator | Symbol | Example | Result |
|---|---|---|---|
| Addition | `+` | `10 + 3` | `13` |
| Subtraction | `-` | `10 - 3` | `7` |
| Multiplication | `*` | `10 * 3` | `30` |
| Division | `/` | `10 / 3` | `3.333... (always float)` |
| Floor division | `//` | `10 // 3` | `3 (drops decimal)` |
| Modulus | `%` | `10 % 3` | `1 (remainder)` |
| Exponent | `**` | `2 ** 10` | `1024` |

> 💡 `//` and `%` are used constantly in DSA — e.g. checking even/odd: `n % 2 == 0`

````python
# ── 5.1 Arithmetic — Run and observe every output ──────────
a, b = 10, 3

print(f'{a} + {b}  = {a + b}')     # 13
print(f'{a} - {b}  = {a - b}')     # 7
print(f'{a} * {b}  = {a * b}')     # 30
print(f'{a} / {b}  = {a / b}')     # 3.333... (always float!)
print(f'{a} // {b} = {a // b}')    # 3  (floor — drops decimal)
print(f'{a} % {b}  = {a % b}')     # 1  (remainder)
print(f'2 ** 10 = {2 ** 10}')      # 1024

# Common use of modulus — even or odd?
num = 17
print(f'{num} is {"even" if num % 2 == 0 else "odd"}')
````

````python
# ── 5.2 Comparison Operators — always return True or False ──
x, y = 7, 3

print(x == y)    # False — equal to
print(x != y)    # True  — not equal
print(x > y)     # True  — greater than
print(x < y)     # False — less than
print(x >= 7)    # True  — greater or equal
print(x <= 6)    # False — less or equal
````

````python
# ── 5.3 Logical Operators ───────────────────────────────────
age   = 20
marks = 85

# and → both must be True
print(age >= 18 and marks >= 60)    # True  — eligible

# or  → at least one must be True
print(age < 18 or marks >= 60)      # True  — marks saves it

# not → flips the boolean
print(not True)                      # False
print(not (age < 18))               # True  — age is NOT < 18
````

````python
# ── 5.4 Assignment Operators (shortcuts) ────────────────────
x = 10
print(f'Start : {x}')

x += 5;  print(f'+= 5  → {x}')   # 15
x -= 3;  print(f'-= 3  → {x}')   # 12
x *= 2;  print(f'*= 2  → {x}')   # 24
x //= 4; print(f'//= 4 → {x}')   # 6
x **= 2; print(f'**= 2 → {x}')   # 36
x %= 10; print(f'%= 10 → {x}')   # 6
````

## 6. Operator Precedence (BODMAS for Python)

When multiple operators appear, Python evaluates in this order:

| Priority | Operators | Note |
|---|---|---|
| 1 — Highest | `**` | Right to left |
| 2 | `*, /, //, %` | Left to right |
| 3 | `+, -` | Left to right |
| 4 | `==, !=, >, <, >=, <=` | Comparison |
| 5 | `not` | Logical NOT |
| 6 | `and` | Logical AND |
| 7 — Lowest | `or` | Logical OR |

> 💡 **When in doubt, use parentheses** `( )` — they always win and make code readable.

````python
# ── 6.1 Precedence — predict before running! ────────────────

# What is the output? Think first, then run.
result1 = 2 + 3 * 4
print('2 + 3 * 4 =', result1)        # 14, not 20 (* before +)

result2 = (2 + 3) * 4
print('(2 + 3) * 4 =', result2)      # 20 (parentheses first)

result3 = 2 ** 3 ** 2
print('2 ** 3 ** 2 =', result3)      # 512 = 2**(3**2) = 2**9

result4 = 10 / 2 * 5
print('10 / 2 * 5 =', result4)       # 25.0 (left to right)

result5 = 2 + 3 * 4 ** 2 / 8 - 1
print('2 + 3 * 4 ** 2 / 8 - 1 =', result5)  # work this out manually first!
````
